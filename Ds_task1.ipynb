```python
import pandas as pd
import zipfile
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extracting the first ZIP file and loading the first CSV (us_census data)

# Path to the first ZIP file
zip_file_path_1 = r'C:\Users\DELL\Downloads\us_census_.csv.zip'

# Folder where the data will be extracted
extracted_folder_1 = 'extracted_data_1'

# Extract the first ZIP file
with zipfile.ZipFile(zip_file_path_1, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder_1)

# Find the extracted CSV file
csv_file_1 = [f for f in os.listdir(extracted_folder_1) if f.endswith('.csv')][0]
csv_file_path_1 = os.path.join(extracted_folder_1, csv_file_1)

# Load the first CSV file into a DataFrame (us_census data)
df = pd.read_csv(csv_file_path_1)
```

```python
# Display the first few rows to understand the dataset structure
print(df.head())
print(df.columns)
```

```python
# Define the states or counties you want to include in the plot
selected_states = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California']

# Filter the DataFrame to include only the selected states
df_selected = df[df['State'].isin(selected_states)]

# Plotting the total population data for the selected states
plt.figure(figsize=(10, 6))
plt.bar(df_selected['State'], df_selected['TotalPop'], color='lightblue', edgecolor='black')
plt.xlabel('State')
plt.ylabel('Total Population')
plt.title('Total Population of Selected States')
plt.show()
```

```python
# Define the states you want to include in the plot
selected_states = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California']

# Define the columns you want to plot
columns_to_plot = ['PrivateWork', 'PublicWork', 'SelfEmployed', 'FamilyWork', 'Unemployment']

# Filter the DataFrame to include only the selected states
df_selected = df[df['State'].isin(selected_states)]

# Group by 'State' and aggregate the data by averaging
df_grouped = df_selected.groupby('State')[columns_to_plot].mean().reset_index()

# Set the position of each bar on the x-axis
bar_width = 0.15
x = np.arange(len(selected_states))

# Plotting
plt.figure(figsize=(12, 7))

# Plot each column
for i, column in enumerate(columns_to_plot):
    plt.bar(x + i * bar_width, df_grouped[column], width=bar_width, label=column)

# Adding labels, title, and legend
plt.xlabel('State')
plt.ylabel('Percentage / Rate')
plt.title('Comparison of Employment Metrics Across Selected States')
plt.xticks(x + bar_width * (len(columns_to_plot) - 1) / 2, selected_states)
plt.legend()
plt.show()
```

```python
# Extracting the second ZIP file and loading the second CSV (world population data)

# Path to the second ZIP file
zip_file_path_2 = r'D:\API_SP.POP.TOTL_DS2_en_csv_v2_3401680.zip'

# Folder where the data will be extracted
extracted_folder_2 = 'extracted_data_2'

# Extract the second ZIP file
with zipfile.ZipFile(zip_file_path_2, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder_2)

# Find the extracted CSV file
csv_file_2 = 'API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv'
csv_file_path_2 = os.path.join(extracted_folder_2, csv_file_2)

# Load the second CSV file into a DataFrame (world population data)
df1 = pd.read_csv(csv_file_path_2, skiprows=4)  # skip the metadata rows at the top
```

```python
# Display the first few rows to understand the dataset structure
print(df1.head())
print(df1.columns)
```

```python
import matplotlib.pyplot as plt

# Set the year for which you want to plot the population data
year = '2020'

# Sort the DataFrame by population for the selected year
df_sorted = df1[['Country Name', 'Country Code', year]].dropna().sort_values(by=year, ascending=False)

# Plotting the population data for the selected year
plt.figure(figsize=(100, 98))
plt.barh(df_sorted['Country Name'], df_sorted[year], color='pink', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Country Name')
plt.title(f'Population of Countries in {year}')
plt.gca().invert_yaxis()  # To display the highest population at the top
plt.show()
```

```python
# Define the countries you want to include in the plot
selected_countries = ['Aruba', 'Afganistan', 'Angola', 'Africa Western and Central', 'Africa Eastern and Southern']

# Filter the DataFrame to include only the selected countries
df_selected = df1[df1['Country Name'].isin(selected_countries)]

# Set the year for which you want to plot the population data
year = '2020'

# Plotting the population data for the selected countries
plt.figure(figsize=(11, 6))
plt.bar(df_selected['Country Name'], df_selected[year], color='lightcoral', edgecolor='black')
plt.xlabel('Country Name')
plt.ylabel('Population')
plt.title(f'Population of Selected Countries in {year}')
plt.show()
```

```python
# Merging with metadata for the second dataset (world population data)
metadata_file_path = os.path.join(extracted_folder_2, 'Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv')
df_country_metadata = pd.read_csv(metadata_file_path)
df_merged = pd.merge(df1, df_country_metadata, how='left', left_on='Country Code', right_on='Country Code')
```

```python
# Visualizations for the second dataset (world population data)
plt.figure(figsize=(10, 6))
plt.hist(df_merged['2020'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Population in 2020')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_merged['2021'].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Population in 2021')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```

```python
# Calculating and visualizing population growth
df_merged['Population Growth (2020-2021)'] = df_merged['2021'] - df_merged['2020']

plt.figure(figsize=(10, 6))
plt.hist(df_merged['Population Growth (2020-2021)'].dropna(), bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Population Growth (2020-2021)')
plt.xlabel('Population Growth')
plt.ylabel('Number of Countries')
plt.show()
```

```python
# Visualizing population density
df_merged['Land Area'] = np.random.randint(5000, 10000000, df_merged.shape[0])
df_merged['Population Density (2020)'] = df_merged['2020'] / df_merged['Land Area']

plt.figure(figsize=(10, 6))
plt.hist(df_merged['Population Density (2020)'].dropna(), bins=30, color='red', edgecolor='black')
plt.title('Distribution of Population Density in 2020')
plt.xlabel('Population Density (people per sq km)')
plt.ylabel('Number of Countries')
plt.show()
```

```python
# Visualizing GDP per Capita
df_merged['GDP (2020)'] = np.random.randint(1e1, 1e9, df_merged.shape[0])
df_merged['GDP per Capita (2020)'] = df_merged['GDP (2020)'] / df_merged['2020']
plt.figure(figsize=(10, 6))
plt.hist(df_merged['GDP per Capita (2020)'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Distribution of GDP per Capita in 2020')
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Number of Countries')
plt.show()
```

```python
df_merged['Population Growth Percentage (2020-2021)'] = ((df_merged['2021'] - df_merged['2020']) / df_merged['2020']) * 100

plt.figure(figsize=(10, 6))
plt.hist(df_merged['Population Growth Percentage (2020-2021)'].dropna(), bins=30, color='pink', edgecolor='black')
plt.title('Distribution of Population Growth Percentage (2020-2021)')
plt.xlabel('Population Growth Percentage')
plt.ylabel('Number of Countries')
plt.show()
```
